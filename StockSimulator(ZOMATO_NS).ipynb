```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.displot(returns, stat = "density", kde = True)
plt.show()
```

```python
from scipy.stats import norm
fit = norm.fit(returns)
sns.displot(returns, kde = True)
plt.plot(np.arange(-0.12,0.12,0.001), norm.pdf(np.arange(-0.12,0.12,0.001), fit[0], fit[1]))
plt.show()
```

From the graph above, it is clear that the distribution follows almost normal distribution. Therefore, we will take the lower_bound and upper_bound for the Monte-Carlo Simulation accordingly. 

```python

lower_bound = norm.ppf(0.25)
upper_bound = norm.ppf(0.75)
mean = returns.mean()
std_dev = returns.std()
np.random.seed(1)
n = np.random.normal(size = (23,10))
rows = n.shape[0]
col = n.shape[1]
for i in  range(rows):
  for j in range(col):
    if n[i][j] > upper_bound:
      n[i][j] = upper_bound
    elif n[i][j] < lower_bound:
      n[i][j] = lower_bound
    n[i][j] = std_dev*n[i][j] + mean
```

```python
new_price = []
prev_price = price.at['2023-01-17']
prev_price
```

```python
for i in range(rows):
  a = []
  for j in range(col):
    a.append(prev_price*(1+n[i][j]))
  prev_price = prev_price*(1+n[i][j])
  new_price.append(a)
```
